```python
import pandas as pd
import os
import sqlite3

db = sqlite3.connect("Proj.db")
df = pd.read_csv('CauseOfDeath.csv')

# Create Main Table
df.to_sql("CauseOfDeath", db, if_exists="replace")
# Create suicides by gender table
df.to_sql("Suicide", db, if_exists="replace")

# Function To Run Queries
def runQuery(query):
    return pd.read_sql_query(query, db)
```

```python
# Display data from Main Table
query = """
SELECT * 
FROM CauseOfDeath;
"""
runQuery(query)
```

```python
#Query Suicide Table 

querySuicideByGender = """
SELECT CASE [sex]                          -- Value Manipulation for Project 3
         WHEN 'Males' THEN 'Male' 
         WHEN 'Females' THEN 'Female' 
       END         AS 'Sex' 
       ,Sum(value) AS "Number Of Deaths"   -- Aggregation for Project 3
       ,Min(time)  AS "Date Range Start" 
       ,Max(time)  AS "Date Range End" 
       ,[geo]      AS "Location" 
       ,[icd10]    AS "Cause Of Death" 
FROM   suicide 
WHERE  [cause of death] LIKE "intentional%" 
GROUP  BY [sex] 
"""

runQuery(querySuicideByGender)
```
